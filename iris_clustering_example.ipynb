```python
import pandas as pd
import sklearn
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(context='paper')
%matplotlib inline
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
data = np.append(X, np.reshape(y, (len(y), 1)), 1)
```

```python
df = pd.DataFrame(data, columns=['sl', 'sw', 'pl', 'pw', 'label'])
```

```python
sns.pairplot(df, vars=df.columns[:4], hue='label')
```

```python
df2 = df[['pl', 'pw', 'label']]
df2.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN().fit_predict(X[:,2:])
```

```python
dbscan
```

```python
idx = list(dbscan).index(1)
plt.scatter(X[:,2:3][:idx], X[:,3:4][:idx],color='red')
plt.scatter(X[:,2:3][idx:], X[:,3:4][idx:],color='blue')
plt.show()
```
